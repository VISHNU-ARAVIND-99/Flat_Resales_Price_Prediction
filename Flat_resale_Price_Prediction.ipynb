```python
df = pd.concat([d1, d2, d3, d4, d5])
df.reset_index(drop=True, inplace=True)
df
```

```python
graph = df.groupby("flat_type").agg({"flat_type": "first", "resale_price": "mean"}).sort_values(by="resale_price")
graph["flat_type"].value_counts()
graph.plot(kind='bar', x='flat_type', y='resale_price', figsize=(10, 6))
plt.title('Average Resale Price by Flat Type')
plt.xlabel('Flat Type')
plt.ylabel('Average Resale Price')
plt.xticks(rotation=45)
plt.show()
```

```python
ordinal_mapping = {'1 ROOM': 1,
                   '2 ROOM': 2,
                   '3 ROOM': 3,
                   '4 ROOM': 4,
                   '5 ROOM': 5,
                   'EXECUTIVE': 6,
                   'MULTI GENERATION': 7,
                   'MULTI-GENERATION': 7} 
df["flat_type"] = df["flat_type"].map(ordinal_mapping)
```

```python
graph = df.groupby(["storey_range"]).agg({"storey_range": "first", "resale_price": "mean"}).sort_values(by="resale_price")
graph.plot(kind='bar', x='storey_range', y='resale_price', figsize=(10, 6))
plt.title('Average Resale Price by storey_range')
plt.xlabel('storey range')
plt.ylabel('Average Resale Price')
plt.xticks(rotation=70)
plt.show()
```

```python
a = df['storey_range'].str.split(' TO ').str[0].astype(int)
b = df['storey_range'].str.split(' TO ').str[1].astype(int)
df["storey_range_upper_limit"] = a
df["storey_range_lower_limit"] = b
df.drop(columns="storey_range", inplace=True)
```

```python
plt.figure(figsize=(15, 7))
bins = np.arange(df['floor_area_sqm'].min(), df['floor_area_sqm'].max(), 15)  # Corrected
plt.scatter(df['floor_area_sqm'], df['resale_price'])
average = [df[(df['floor_area_sqm'] >= b) & (df['floor_area_sqm'] < b + 15)]['resale_price'].mean() for b in bins]
plt.plot(bins + 7.5, average, color='r', linestyle='--', marker='o', label='resale_price')
plt.xlabel('Floor Area (sqm)')
plt.ylabel('Resale Price')
plt.title('Scatter Plot with Average Resale Price by Floor Area in sqm')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df["month"] = pd.to_datetime(df["month"], format='%Y-%m')
df["lease_commence_date"] = pd.to_datetime(df["lease_commence_date"], format='%Y')
f = df['month'] - df['lease_commence_date']
w = f/pd.Timedelta(days=365.25)
df["year_old"] = w
df = df[df["year_old"] >= 0]
```

```python
df["lease_remaining_year"] = 100 - df["year_old"]
df
```

```python
df.drop(columns="lease_commence_date", inplace=True)
df["resale_year"] = df["month"].dt.year
df["resale_month"] = df["month"].dt.month
df.drop(columns="month", inplace=True)
```

```python
#lease_remaining_year ==> -1.00
# storey_range_upper_limit ==> 1.00
y = df["resale_price"]
x = df.drop(columns=["lease_remaining_year", "storey_range_upper_limit", "resale_price"])
```

```python
x.head()
```

```python
import category_encoders as ce

encoder = ce.BinaryEncoder(cols=['flat_model', 'town'])
x = encoder.fit_transform(x)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=45)
    
model = RandomForestRegressor().fit(x_train, y_train)
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
accuracy = {
            'R2' : r2,
            'MSE' : mse,
            'RMSE'  : rmse,
            'MAE' : mae,
           }
print(accuracy)
```
